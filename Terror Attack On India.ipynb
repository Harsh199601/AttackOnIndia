```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv("india_data.csv")
endDataset=dataset.tail(n=4351)
Xyear=endDataset['year']
Ytarget=endDataset['weapon']
plt.bar(Xyear,Ytarget,color='Blue')
plt.title('Weapon Used For the Attack!')
plt.xlabel('Years')
plt.ylabel('Weapon Used ')
plt.rcParams["figure.figsize"] = (15, 9)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv("india_data.csv")
endDataset=dataset.tail(n=4351)
newDataSet=endDataset[endDataset['target']=='Government (Diplomatic)']
injury=newDataSet['total']
state=newDataSet['state']
plt.pie(injury,labels=state,shadow=True,explode=(0.2,0,0),autopct='%0.1f%%')
plt.legend()
plt.show()
```
